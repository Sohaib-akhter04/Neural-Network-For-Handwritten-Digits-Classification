```python
import tensorflow as tf
from tensorflow import keras 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

This code is using the Keras deep learning library to load the MNIST dataset. MNIST is a classic dataset of handwritten digits that is often used as a benchmark in image recognition tasks.

```python
x_train=x_train/255
x_test=x_test/255 #scaling from 0 to 1 to increase the accuracy
```

```python
x_train[0]
```

```python
X_train_flattened=x_train.reshape(len(x_train),28*28)
X_test_flattened=x_test.reshape(len(x_test),28*28)
```

```python
X_test_flattened.shape #2d into 1D array
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
]) #input shape is 784 input layer and output is 10(0-9) output layer 
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']

)#optimizer allows you to train efficienttly for backward propogation
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
plt.matshow(x_test[0])
```

```python
y_predicted=model.predict(X_test_flattened)  #predicting via test data
y_predicted[0]
```

```python
np.argmax(y_predicted[0]) #max from the 1x10 array stored in the 0 index
```

```python
y_predicted[1]
np.argmax(y_predicted[1])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels) #kyke hume labels pass krne rather than predicted values which is a values of each digit
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#MODEL WITH HIDDEN LAYERS
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')

]) #input shape is 784 input layer and output is 10(0-9) output layer 
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']

)#optimizer allows you to train efficienttly for backward propogation
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted=model.predict(X_test_flattened)  
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels) #kyke hume labels pass krne rather than predicted values which is a values of each digit

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#MODEL WITH HIDDEN LAYERS
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')

]) #input shape is 784 input layer and output is 10(0-9) output layer 
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']

)#optimizer allows you to train efficienttly for backward propogation
model.fit(x_train,y_train,epochs=5) #without flattening array
```
